```python
FILE_PATH = "../data/frontend_resume.pdf"
def show_metadata(docs):
    if docs:
        print("[metadata]")
        print(list(docs[0].metadata.keys()))
        print("\n[examples]")
        max_key_length = max(len(k) for k in docs[0].metadata.keys())
        for k, v in docs[0].metadata.items():
            print(f"{k:<{max_key_length}} : {v}")
```

```python
from langchain_community.document_loaders import PyPDFLoader

# 파일 경로 설정
loader = PyPDFLoader(FILE_PATH)

# PDF 로더 초기화
docs = loader.load()

# 문서의 내용 출력
for doc in docs:
    resume_text = doc.page_content
# print(docs[1].page_content[:300])
```

```python
resume_text
```

```python
# 유사도 계산
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

# 이력서 텍스트 임베딩
resume_embedding = model.encode(resume_text)

# 질문 텍스트 임베딩
questions = job_question("프론트엔드")["질문"].tolist()
question_embeddings = model.encode(questions)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity([resume_embedding], question_embeddings)
most_similar_index = similarities.argmax()
most_similar_question = questions[most_similar_index]

print("가장 유사한 질문:", most_similar_question)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def get_top_n_similar_questions(resume_text, questions, n=5):
    # 이력서와 질문 임베딩
    resume_embedding = model.encode(resume_text)
    question_embeddings = model.encode(questions)

    # 유사도 계산
    similarities = cosine_similarity([resume_embedding], question_embeddings)[0]

    # 유사도 상위 n개의 인덱스 추출
    top_n_indices = np.argsort(similarities)[-n:][::-1]  # 유사도 높은 순으로 정렬

    # 결과 반환 (질문과 유사도 점수)
    top_n_questions = [(questions[i], similarities[i]) for i in top_n_indices]
    return top_n_questions
```

```python
# 3. 유사도 계산 및 결과 출력
top_questions = get_top_n_similar_questions(resume_text, questions, n=7)
print("\n가장 유사한 질문 상위 5개:")
for i, (question, similarity) in enumerate(top_questions, start=1):
    print(f"{i}. {question} (유사도: {similarity:.2f})")
```
